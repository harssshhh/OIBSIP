# Importing the data set into pandas data frame

```python

df = pd.read_csv('/kaggle/input/iriscsv/Iris.csv')
```

```python
df
```

```python
# dropping the id column
df = df.drop(columns=['Id'])
df.head()
```

```python
# checking all the unique values
df.Species.value_counts()
```

```python
df.shape
```

```python
# Converting species types into numerical format by replacing each one of them with a different integer
df["Species"].replace({"Iris-setosa": 2, "Iris-versicolor": 3, "Iris-virginica": 4}, inplace = True)
df.head()
```

```python
X=pd.DataFrame(df,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
Y=df.Species.values.reshape(-1,1)
```

```python
X[0:5] # representing the input attributes
```

```python
Y # representing the target variable species
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import warnings
```

# Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
k = 7

kn = KNeighborsClassifier(k)

kn.fit(X_train,Y_train)

Y_pred = kn.predict(X_test)
```

# Checking the accuracy score of classification model

```python
metrics.accuracy_score(Y_test,Y_pred)*100
```
